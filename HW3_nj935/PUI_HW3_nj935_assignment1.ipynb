```python
# Import the necessary libraries

import numpy as np
import pylab as pl
import seaborn
%pylab inline

# Set variables to use with the probability distributions below

mymean=100
df = mymean
std=4
N=30
mybins=30
mysize = np.linspace(20,1000,100)
md={}
```

```python
# Normal

md['normal']={}

for n in mysize:
    md['normal'][n] = np.random.normal(mymean, 2, size = n)
    
md['normal']['means'] = {}
ax_mu_n = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in md['normal']:
    if not type(nn) == str:
        md['normal']['means'][nn] = md['normal'][nn].mean()
        #and plot it
        ax_mu_n.plot(nn, md['normal']['means'][nn], 'o')
        ax_mu_n.set_xlabel('sample size', fontsize=18)
        ax_mu_n.set_ylabel('sample mean', fontsize=18)
        ax_mu_n.set_title('Fig 1: 100 random samples following the normal distribution (sample size 20 < N < 2000)', fontsize=18)
        ax_mu_n.plot([min(mysize), max(mysize)], [df, df], 'k')
```

```python
allmeans = list(md['normal']['means'].values())

pl.figure(figsize=(6, 6))
pl.hist(allmeans,bins=30)
pl.title('Histogram of mean values for normal random samples', fontsize=14)
pl.xlabel('sample mean', fontsize = 14)
pl.ylabel('N', fontsize = 14)
```

### Explanation of central limit theorem

The histograms for each set of samples show the frequency of each unique value. The binomial, poisson, chi-square and logistic samples all follow these probability distributions. However, repeat each one of them 100 times, and we see that the distribution of their sample means is normally distributed.

This is due to the *central limit theorem*, and is a very helpful property, because it means
we can analyze phenomena that follow other distributions using the convenient methods
associated with the normal distribution

```python
# Plot a histogram of the means from each random sample across the 5 sets

reallyallmeans = (md['normal']['means'].values()+(md['poisson']['means'].values())+(md['chisq']['means'].values())+(md['binomial']['means'].values())+(md['logistic']['means'].values()))

pl.figure(figsize=(10, 10))
pl.hist(reallyallmeans,bins=30)
pl.title('Histogram of mean values for all 5 random samples combined', fontsize=18)
pl.xlabel('sample mean', fontsize = 14)
pl.ylabel('N', fontsize = 14)
```
