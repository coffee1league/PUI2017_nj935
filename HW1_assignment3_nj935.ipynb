# Create a *reproducible research* chunk of code
This notebook gives my code and plot for assignment 3. This was fun!

```python
# Import libraries

import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

# Set random seed

np.random.seed(0)

# Generate 50x2x100 array
reprandall = np.zeros((50, 2, 100))

# For loop to populate array with random values
# Give each distribution a different randomly generated mean and variance
for q in range(50):
        mu = np.random.randn(1)
        sigma = np.random.randn(1)
        reprandall[q] = sigma * np.random.randn(2,100) + mu

# Plot the figure
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111)
x = reprandall[:,0,:]
y = reprandall[:,1,:]
colors = np.random.randn(5000)
plt.scatter(x,y,c=colors,alpha=0.5)

# Add title and labels
plt.title("Figure 1: Reproducible random plot")
plt.xlabel("X axis - mean % f, st dev % f" % (x.mean(),x.std()))
plt.ylabel("Y axis - mean %.4f, st dev %.4f" % (y.mean(),y.std()))
```
